# Image Convolution

藉由 image convolution 的操作來說明 Convolutional Neural Networks 的 feature extraction 部分的運作原理。
https://en.wikipedia.org/wiki/Sobel_operator

```python
%matplotlib inline
import numpy as np
from scipy import ndimage
from PIL import Image
import matplotlib.pyplot as plt
```

### ---------------------------------------------------------------- Lena ---------------------------------------------------------------

## 讀取圖形

```python
# 讀取影像
lena = Image.open('lena.jpg')
```

```python
lena.size
```

```python
lena.mode
```

```python
# 將彩色影像轉換成灰階影像
lena_gray = lena.convert('L')
```

```python
lena_gray.mode
```

```python
# 改成np.array
lena_array = np.array(lena_gray)
```

```python
plt.imshow(lena_array, cmap='gray');
```

## TODO: 將 img_array 做一點處理，防止 data overflow...

圖型為 RGB :3X3 陣列分開處理。

```python
# 型態為uint8，在相乘時只到255
lena_array.dtype
```

```python
lena_array=lena_array.astype('int64')
```

## 試試看 sobel filter 的效果

```python
sobelx = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]], dtype = np.float)
```

```python
sobely = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]], dtype = np.float)
```

```python
# Convolve
# X
outx = ndimage.convolve(lena_array, sobelx).astype('float')
```

```python
plt.imshow(outx, cmap='gray');
```

```python
# Y
outy = ndimage.convolve(lena_array, sobely).astype('float')
```

```python
plt.imshow(outy, cmap='gray');
```

```python
# X平方、Y平方相加，開根號
out = np.sqrt(outx*outx+outy*outy)
```

```python
plt.imshow(out, cmap='gray');
```
